```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
city = ['Berlin', 'Frankfurt', 'Munich', 'Braunschweig', 'Hanover', 'Kassel','Karlsruhe', 'D%C3%BCsseldorf']
city_pop_num = ['18', '21', '21', '19', '19', '18', '20', '19']
urban_pop_num = ['19','23','23',None,None,None,None,'21']
area_num = ['15','18','18','16','16','15','17','16']
ele_num = ['16','19','19','17','17','16','18','17']
print(len(city), len(city_pop_num), len(urban_pop_num), len(area_num), len(ele_num))
for index in range(len(city)):
    print(index, city[index])
```

```python
def scrap_data(soup, city_pop_num, urban_pop_num, area_num, ele_num):
    city_pop = None
    urban_pop = None
    area = None
    elevation = None
    if (city_pop_num != None) : 
        city_pop = soup.select(f'#mw-content-text > div.mw-parser-output > table.infobox.ib-settlement.vcard > tbody > tr:nth-child({city_pop_num}) > td')[0].get_text() 
    if (urban_pop_num != None) : 
        urban_pop = soup.select(f'#mw-content-text > div.mw-parser-output > table.infobox.ib-settlement.vcard > tbody > tr:nth-child({urban_pop_num}) > td')[0].get_text()
    if (area_num!= None) :
        area= soup.select(f'#mw-content-text > div.mw-parser-output > table.infobox.ib-settlement.vcard > tbody > tr:nth-child({area_num}) > td')[0].get_text() 
    if (ele_num!= None):
        elevation = soup.select(f'#mw-content-text > div.mw-parser-output > table.infobox.ib-settlement.vcard > tbody > tr:nth-child({ele_num}) > td')[0].get_text()

    return city_pop, urban_pop, area, elevation
```

```python
city_pop = [] 
urban_pop = []
area = [] 
elevation = [] 
latitude = []
longitude = []
for index in range(len(city)):
    #print(city[index])
    wiki = "https://en.wikipedia.org/wiki/" + city[index]
    # print(wiki)
    response = requests.get(wiki)
    # print(f"Wiki: {response.status_code}")
    soup = BeautifulSoup(response.content, "html.parser")
    # print(soup)
    pop_city, pop_urban, area_var, elevation_var = scrap_data(soup, city_pop_num[index], urban_pop_num[index], area_num[index], ele_num[index])
    #print(pop_city, pop_urban, area_var, elevation_var)

    latlon = soup.find("span", class_="geo").get_text()
    lat,lon = latlon.split(';')
    #print(lat, lon)
    
    
    city_pop.append(pop_city)
    urban_pop.append(pop_urban)
    area.append(area_var)
    elevation.append(elevation_var)
    latitude.append(lat)
    longitude.append(lon)

    # print(city_pop,urban_pop,area,elevation)
    # print(latitude,longitude)


city_df = pd.DataFrame({
        'city': city,
        'city_population' : city_pop,
        'urban_population': urban_pop,
        'area': area,
        'altitude': elevation,
        'latitude': latitude,
        'longitude': longitude,
        'airport_IATA': ['BER', 'FRA', 'MUC', 'BWE', 'HAJ', 'KSF', 'FKB', 'DUS']
}
)

city_df   
```
