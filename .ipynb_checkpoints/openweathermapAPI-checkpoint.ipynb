We can also check weather information by the name of  a particular city!
Now, I'll create a dataframe with 7 days weather information for these cities: Berlin', 'Frankfurt am Main', 'Munich', 'Braunschweig', 'Hannover', 'Kassel', 'Karlsruhe', 'Düsseldorf' 

```python
german_cities = ['Berlin', 'Frankfurt am Main', 'Munich', 'Braunschweig', 'Hannover', 'Kassel', 'Karlsruhe', 'Düsseldorf']

temp_city = []
temp_date = []
day_temp = []
max_temp = []
min_temp = []
night_temp = []
status = []
detailed_status = []
precipitation_probability = []

for city in german_cities:
    city_lat = reg.geopoints_for(city)[0].lat
    city_lon = reg.geopoints_for(city)[0].lon
    one_call = mgr.one_call(lat=city_lat, lon=city_lon)
    for i in range(7):
        temp_city.append(city)
        temp_date.append(one_call.forecast_daily[i].reference_time('iso')[:10])
        day_temp.append(one_call.forecast_daily[i].temperature('celsius')['day'])
        max_temp.append(one_call.forecast_daily[i].temperature('celsius')['max'])
        min_temp.append(one_call.forecast_daily[i].temperature('celsius')['min'])
        night_temp.append(one_call.forecast_daily[i].temperature('celsius')['night'])
        status.append(one_call.forecast_daily[i].status)
        detailed_status.append(one_call.forecast_daily[i].detailed_status)
        precipitation_probability.append(one_call.forecast_daily[i].precipitation_probability)
```

```python
weather_df = pd.DataFrame({"city": temp_city,
                            "date": temp_date,
                            "day_temp": day_temp,
                            "max_temp": max_temp,
                            "max_temp": min_temp,
                            "night_temp": night_temp,
                            "status": status,
                            "detailed_status": detailed_status,
                            "precipitation_probability": precipitation_probability
                           })
weather_df
```
